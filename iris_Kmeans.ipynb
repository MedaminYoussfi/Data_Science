```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
y = iris.target
X.shape
```

```python
# Tracé du nuage de points initial
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Nuage des points  initiiales avant le groupement des iris')
plt.show()
```

```python
# Creation d'un objet KMeans avec k clusters
k=3
kmeans = KMeans(n_clusters=k)

# Entraînement du modèle
kmeans.fit(X)

# Obtention des affectations de cluster pour chaque échantillon
labels = kmeans.labels_

# Obtention des centres de cluster
centers = kmeans.cluster_centers_


centers
```

```python
# Tracé du nuage de points
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means où (K={k})')
plt.show()
```

```python
# Tracé du nuage de points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means où (K={k})')
plt.show()
```

```python
# Tracé du nuage de points avant le groupement des iris
plt.subplot(1, 2, 1)  # Spécifier la position du subl*plot
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means ')

# Tracé du nuage de points avec K-means (k clusters)
plt.subplot(1, 2, 2)  # Specify the subplot position
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x')
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'Nuage de points avec K-means')
plt.show()
```

```python
# Liste pour stocker les valeurs d'inertie
inertie = []

# Nb Clusters à tester
k_values = range(1, 15)

# Calcul inertie pour chaque valeur de k
for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertie.append(kmeans.inertia_)

# Tracé de la courbe du coude
plt.plot(k_values, inertie, 'bx-')
plt.xlabel('Nombre de clusters (k)')
plt.ylabel('Inertie')
plt.title('Courbe du coude')
plt.show()
```

```python
# Calcul de l'inertie pour k=3
kmeans_3 = KMeans(n_clusters=3)
kmeans_3.fit(X)
inertie_3 = kmeans_3.inertia_
print("Inertie pour k=3 :", inertie_3)

# Calcul de l'inertie pour k=4
kmeans_4 = KMeans(n_clusters=4)
kmeans_4.fit(X)
inertie_4 = kmeans_4.inertia_
print("Inertie pour k=4 :", inertie_4)
```
