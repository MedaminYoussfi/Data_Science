```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.lib.function_base import blackman
```

```python
x = np.linspace(0, 2, 10)
y = x**2
plt.plot(x, y)
plt.show()
```

```python
plt.scatter(x, y)
plt.show()
```

```python
plt.plot(x, y, c='b', lw=10, ls='--', alpha=0.5)
plt.show()
```

```python
x = np.linspace(0, 2, 10)
print(x)
plt.figure()
plt.plot(x, x**2, label='quadratiique') # courbe 1
plt.plot(x, x**3, label='cubique') # courbe 2
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.savefig('figure.png') # sauvegarder la figure
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(x, x*2, c='r')
plt.subplot(2, 1, 2)

plt.plot(x, x**2, c='b')
```

```python
fig, ax = plt.subplots(2, 2, sharex=True) # partage le meme axe pour les subplots
ax[0,0].plot(x, y)
ax[0,1].plot(x, np.sin(x),c='r')
ax[1,0].plot(x, np.cos(x), c='g')
ax[1,1].plot(x, np.cos(x))
plt.show()
```

```python
def graphique(data):
    n = len(data)
    plt.figure(figsize=(12, 20))

    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)

    plt.show()
```
